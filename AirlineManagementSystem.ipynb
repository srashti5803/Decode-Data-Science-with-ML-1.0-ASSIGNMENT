```python
# 10.
# Write a code for Airline Reservation System using OO4S3
# & Create a Flight 'lass that has attributes su'h as flight number, departure and arrival airports, departure and 
# arrival times, and available seats (private)
# & Implement methods to book a seat, 'an'el a reservation, and get the remaining available seats
# & Use en'apsulation to hide the flight's unique identifi'ation number
# & Inherit from the Flight 'lass to 'reate a Domesti'Flight 'lass and an InternationalFlight 'lass, ea'h with their 
# own spe'ifi' attributes and methods


class Flight:
    def __init__(self, flight_number, departure_airport, arrival_airport, departure_time, arrival_time, total_seats):
        self.__flight_number = flight_number
        self.__departure_airport = departure_airport
        self.__arrival_airport = arrival_airport
        self.__departure_time = departure_time
        self.__arrival_time = arrival_time
        self.__total_seats = total_seats
        self.__available_seats = total_seats

    def book_seat(self):
        if self.__available_seats > 0:
            self.__available_seats -= 1
            print(f"Seat booked on Flight {self.__flight_number}.")
        else:
            print(f"No available seats left on Flight {self.__flight_number}.")

    def cancel_booking(self):
        if self.__available_seats < self.__total_seats:
            self.__available_seats += 1
            print(f"Booking cancelled on Flight {self.__flight_number}.")
        else:
            print(f"No bookings to cancel on Flight {self.__flight_number}.")

    def get_available_seats(self):
        return self.__available_seats

    def get_flight_number(self):
        return self.__flight_number

    def get_departure_airport(self):
        return self.__departure_airport

    def get_arrival_airport(self):
        return self.__arrival_airport

    def get_departure_time(self):
        return self.__departure_time

    def get_arrival_time(self):
        return self.__arrival_time

    def display_info(self):
        print(f"Flight Number: {self.__flight_number}")
        print(f"Departure Airport: {self.__departure_airport}")
        print(f"Arrival Airport: {self.__arrival_airport}")
        print(f"Departure Time: {self.__departure_time}")
        print(f"Arrival Time: {self.__arrival_time}")
        print(f"Available Seats: {self.__available_seats}/{self.__total_seats}")
class DomesticFlight(Flight):
    def __init__(self, flight_number, departure_airport, arrival_airport, departure_time, arrival_time, total_seats, baggage_allowance):
        super().__init__(flight_number, departure_airport, arrival_airport, departure_time, arrival_time, total_seats)
        self.__baggage_allowance = baggage_allowance

    def get_baggage_allowance(self):
        return self.__baggage_allowance

    def display_info(self):
        super().display_info()
        print(f"Baggage Allowance: {self.__baggage_allowance} kg")


class InternationalFlight(Flight):
    def __init__(self, flight_number, departure_airport, arrival_airport, departure_time, arrival_time, total_seats, meal_service):
        super().__init__(flight_number, departure_airport, arrival_airport, departure_time, arrival_time, total_seats)
        self.__meal_service = meal_service

    def has_meal_service(self):
        return self.__meal_service

    def display_info(self):
        super().display_info()
        print(f"Meal Service: {'Yes' if self.__meal_service else 'No'}")
# Test the Airline Reservation System
if __name__ == "__main__":
    # Create a DomesticFlight and an InternationalFlight
    domestic_flight = DomesticFlight("DL123", "JFK", "LAX", "08:00", "12:00", 150, 25)
    international_flight = InternationalFlight("AI456", "DEL", "LHR", "14:00", "22:00", 200, True)

    # Test flight operations
    print("Booking Seats:")
    domestic_flight.book_seat()
    domestic_flight.book_seat()
    international_flight.book_seat()

    print("\nCancelling Bookings:")
    domestic_flight.cancel_booking()
    international_flight.cancel_booking()

    print("\nUpdated Flight Information:")
    domestic_flight.display_info()
    print("\n")
    international_flight.display_info()
```
