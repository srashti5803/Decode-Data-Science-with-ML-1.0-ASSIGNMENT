```python
# 9.Write a code for  Event Management System using OO4S3
# & Create an Event 'lass that has attributes su'h as name, date, time, lo'ation, and list of attendees (private)
# & Implement methods to 'reate a new event, add or remove attendees, and get the total number of 
# attendees
# & Use en'apsulation to hide the event's unique identifi'ation number
# & Inherit from the Event 'lass to 'reate a PrivateEvent 'lass and a Publi'Event 'lass, ea'h with their own 
# spe'ifi' attributes and methods



class Event:
    def __init__(self, name, date, time, location):
        self.__name = name
        self.__date = date
        self.__time = time
        self.__location = location
        self.__attendees = []  # List of attendees

    def create_event(self):
        print(f"Event '{self.__name}' created on {self.__date} at {self.__time}, {self.__location}.")

    def add_attendee(self, attendee):
        self.__attendees.append(attendee)
        print(f"{attendee} added to the attendee list.")

    def remove_attendee(self, attendee):
        if attendee in self.__attendees:
            self.__attendees.remove(attendee)
            print(f"{attendee} removed from the attendee list.")
        else:
            print(f"{attendee} is not in the attendee list.")

    def get_total_attendees(self):
        return len(self.__attendees)

    def get_name(self):
        return self.__name

    def get_date(self):
        return self.__date

    def get_time(self):
        return self.__time

    def get_location(self):
        return self.__location

    def display_info(self):
        print(f"Event: {self.__name}")
        print(f"Date: {self.__date}")
        print(f"Time: {self.__time}")
        print(f"Location: {self.__location}")
        print(f"Total Attendees: {self.get_total_attendees()}")

class PrivateEvent(Event):
    def __init__(self, name, date, time, location, host):
        super().__init__(name, date, time, location)
        self.__host = host

    def get_host(self):
        return self.__host

    def display_info(self):
        super().display_info()
        print(f"Host: {self.__host}")


class PublicEvent(Event):
    def __init__(self, name, date, time, location, organizer):
        super().__init__(name, date, time, location)
        self.__organizer = organizer

    def get_organizer(self):
        return self.__organizer

    def display_info(self):
        super().display_info()
        print(f"Organizer: {self.__organizer}")
# Test the Event Management System
if __name__ == "__main__":
    # Create a PrivateEvent and a PublicEvent
    private_event = PrivateEvent("Birthday Party", "2024-07-15", "18:00", "Private Residence", "John Doe")
    public_event = PublicEvent("Community Picnic", "2024-08-20", "12:00", "Central Park", "Local Community Center")

    # Test event operations
    print("Creating Events:")
    private_event.create_event()
    public_event.create_event()

    print("\nAdding Attendees:")
    private_event.add_attendee("Alice")
    private_event.add_attendee("Bob")
    public_event.add_attendee("Charlie")
    public_event.add_attendee("Diana")

    print("\nRemoving Attendees:")
    private_event.remove_attendee("Bob")
    public_event.remove_attendee("Diana")

    print("\nUpdated Event Information:")
    private_event.display_info()
    print("\n")
    public_event.display_info()
```
